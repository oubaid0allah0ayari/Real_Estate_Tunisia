```python
df = pd.read_csv('Data/raw_data.csv')
```

```python
df = df.drop("URL", axis=1)
df.head()
```

```python
# Clean the Location column by removing unwanted text
df['Location'] = df['Location'].str.replace('Se connecter | Immobilier Tunisie | ', '', regex=False)

# Also remove any leading/trailing whitespace and multiple spaces
df['Location'] = df['Location'].str.strip().str.replace(r'\s+', ' ', regex=True)

# Display the cleaned data
print("Location column after cleaning:")
print(df['Location'].head(10))
df.head()
```

```python
# Split the Location column into separate columns using "|" separator
location_split = df['Location'].str.split('|', expand=True)

# Clean each column by removing leading/trailing whitespace
for col in location_split.columns:
    location_split[col] = location_split[col].str.strip()

# Rename the columns with meaningful names
location_columns = []
for i in range(len(location_split.columns)):
    location_columns.append(f'Location_Part_{i+1}')

location_split.columns = location_columns

# Add these new columns to the original dataframe
df = pd.concat([df, location_split], axis=1)

# Drop the original Location column since we now have separate columns
df = df.drop('Location', axis=1)

print("Location data split into separate columns:")
print(location_split.head())
print("\nDataFrame with new location columns:")
df.head()
```

```python
list=df["Details"]
print(list)
```

```python

raw_string=list[1]
pairs = [item.strip().split(": ", 1) for item in raw_string.split("|")]
print (pairs)
```

```python
# Convert pairs to a dictionary for easier handling
details_dict = {}
for pair in pairs:
    if len(pair) == 2:  # Only process pairs that have both key and value
        key, value = pair
        details_dict[key] = value

print("Details as dictionary:")
print(details_dict)

# Now let's apply this to all rows in the Details column
def parse_details(details_string):
    """Parse a details string into a dictionary"""
    if pd.isna(details_string) or details_string == 'nan':
        return {}
    
    try:
        pairs = [item.strip().split(": ", 1) for item in str(details_string).split("|")]
        details_dict = {}
        for pair in pairs:
            if len(pair) == 2:
                key, value = pair
                details_dict[key.strip()] = value.strip()
        return details_dict
    except:
        return {}

# Apply to all rows
df['Details_Dict'] = df['Details'].apply(parse_details)
print("\nFirst few parsed details:")
print(df['Details_Dict'].head())
```

```python
df
```

```python
# Extract Details_Dict into separate columns

# First, get all unique keys across all rows
all_keys = set()
for details_dict in df['Details_Dict']:
    if isinstance(details_dict, dict):
        all_keys.update(details_dict.keys())

print(f"Found {len(all_keys)} unique detail keys:")
# Convert to list without using list() function (since 'list' variable exists)
keys_sample = [key for key in all_keys][:10]
print(keys_sample)  # Show first 10

# Create new columns for each key
for key in all_keys:
    # Clean the key to make it a valid column name
    clean_key = key.replace('é', 'e').replace('è', 'e').replace('à', 'a').replace(' ', '_')
    
    # Extract values for this key, use None if key doesn't exist
    df[clean_key] = df['Details_Dict'].apply(
        lambda x: x.get(key, None) if isinstance(x, dict) else None
    )

print(f"\nAdded {len(all_keys)} new columns from Details_Dict")

# Show the new columns
print("New detail columns added:")
for col in sorted(all_keys):
    clean_col = col.replace('é', 'e').replace('è', 'e').replace('à', 'a').replace(' ', '_')
    print(f"'{col}' -> column '{clean_col}'")

# Display sample of the new structure
print("\nSample of new columns (first 3 rows):")
sorted_keys = sorted([key for key in all_keys])[:5]
detail_cols = [col.replace('é', 'e').replace('è', 'e').replace('à', 'a').replace(' ', '_') for col in sorted_keys]
if detail_cols:
    print(df[detail_cols].head(3))
```
