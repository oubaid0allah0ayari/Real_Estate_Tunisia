```python
# Drop the specified columns that are not needed for price prediction
columns_to_drop = [
    # Appliance/furnishing details (too specific, may not affect price significantly)
    'Chambre_rangement',    # Storage room
    'Refrigerateur',        # Refrigerator
    'Machine_a_laver',      # Washing machine
    'Micro-ondes',          # Microwave
    'Four',                 # Oven
     
    # Building features (may be less important for price prediction)
    'Ascenseur',            # Elevator
    'Concierge',            # Concierge
    'Porte_blindee',        # Security door
    'Fa√ßade_exterieure',    # Exterior facade
    'Salon_europeen',       # European living room
    'Double_vitrage',       # Double glazing
    '√âtage_du_bien',        # Floor level
    'Entre-seul',          # Unclear meaning
    # Additional problematic columns
    '√âtat',                 # Duplicate of 'Etat'
    'Livraison:',          # Formatting issues
    'Surface_de_la_parcelle', # Plot size
    'Surface_exterieure',   # Redundant
    'Surface_habitable',    # Redundant
    'Antenne_parabolique', # Satellite dish
    'Type_du_sol',
    'Securite',
    'Type_de_bien',
    'Type_de_bien',
    'nom appartement',
    'Orientation'
]

print("Columns to drop:")
for col in columns_to_drop:
    if col in df.columns:
        print(f"- {col}")
    else:
        print(f"- {col} (not found in dataset)")

print(f"\nDataset shape before: {df.shape}")
```

```python
# Actually drop the columns
df_cleaned = df.drop(columns=[col for col in columns_to_drop if col in df.columns])

print(f"Dataset shape after dropping columns: {df_cleaned.shape}")
print(f"Dropped {df.shape[1] - df_cleaned.shape[1]} columns")
print(f"Remaining columns: {df_cleaned.shape[1]}")

# Update df to the cleaned version
df = df_cleaned

# Show final column list
print("\nFinal columns in the cleaned dataset:")
print(df.columns.tolist())
```

## üéØ Why These Features?

These 23 features were selected because they:

- **Represent key buyer decision factors** in the Tunisian real estate market
